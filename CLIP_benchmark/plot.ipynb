```python
import json
import numpy as np
import matplotlib.pyplot as plt


f = open('rejection.json')
  
data = json.load(f)

for dataset in data:
    rejection_percentages = np.arange(0. , 1., 0.05)
    non_rejected_accuracies = []
    classification_qualities = []
    rejection_qualities = []
    methods_name = []

    for result in data[dataset]['results']:
        if result not in ['acc1', 'acc5', 'mean_per_class_recall']:
            for method in data[dataset]['results'][result]:
                if method == 'non-rejected-accuracy':
                    non_rejected_accuracies.append((result,data[dataset]['results'][result][method]))
                elif method == 'classification-quality':
                    classification_qualities.append((result,data[dataset]['results'][result][method]))
                elif method == 'rejection-quality':
                    rejection_qualities.append((result,data[dataset]['results'][result][method]))
                

    #print(non_rejected_accuracies)
    for method in non_rejected_accuracies:
        plt.plot(rejection_percentages, method[1], label=method[0])
    
    plt.xlabel('Rejection Percentage')
    plt.ylabel('Non-rejected accuracy (%)')
    plt.title('Non-rejected accuracy on ' + dataset)
    plt.legend()
    plt.show()

    for method in classification_qualities:
        plt.plot(rejection_percentages, method[1], label=method[0])
    
    plt.xlabel('Rejection Percentage')
    plt.ylabel('Classification quality (%)' )
    plt.title('Classification quality on ' + dataset)
    plt.legend()
    plt.show()

    for method in rejection_qualities:
        plt.plot(rejection_percentages, method[1], label=method[0])
    
    plt.xlabel('Rejection Percentage')
    plt.ylabel('Rejection quality')
    plt.title('Rejection quality on ' + dataset)
    plt.legend()
    plt.show()


                # Plot the accuracies against rejection percentages
            #plt.plot(rejection_percentages, accuracies, label=method.__name__ 

    # plt.xlabel('Rejection Percentage')
    # plt.ylabel('Accuracy (%)')
    # plt.title('Accuracy vs. Rejection Percentage')
    # plt.legend()
    # plt.show()
      
```
